```python
xlambda = []
yflux = []
with open('U10.txt','r') as f:
    for line in f:
        xlambda.append(float(line.split()[0]))
        yflux.append(float(line.split()[1]))
        

# create a new plot
p = figure(plot_width=900, plot_height=500, title="U10",active_drag='pan', active_scroll='wheel_zoom')

#Tool to get wavelength
hover2 = HoverTool(
        tooltips=[
            ("(x,y)", "($x{1}, $y)"),
        ]
    )


#plot Lines


##Emission Lines
class line(object):
    def __init__(self,name):
        self.name = name


# add some renderers
p.line(xlambda,yflux)





p.add_tools(hover2)

show(p)
```
